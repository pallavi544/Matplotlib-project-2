## matplotlib

### Plot 

```python
import matplotlib.pyplot as plt
x=df['Date']
y=df['Sales']
plt.plot(x,y,marker='o',ms=5,mfc='yellow')
plt.title('Sales vs Date')
plt.xlabel('Dates')
plt.ylabel('SALES')
plt.show()
           
```

```python
x=df['Date']
y=df['Profit']
plt.plot(x,y,marker='o',ms=5,mfc='red',color='green')
plt.title('Profit vs Date')
plt.xlabel('Dates')
plt.ylabel('profit')
plt.show()
           
```

#### bar

```python
d=df.groupby('Category')['Sales'].mean()
d
```

```python
x=d.index
y=d.values
plt.bar(x,y,color='lightgreen',edgecolor='black')
plt.title('Category and Sales Data')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
x=df['Product']
y=df['Sales']
plt.bar(x,y,color='hotpink',edgecolor='black')
plt.title('Product and Sales Data')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
x=df['Product']
y=df['Profit']
plt.bar(x,y,color='goldenrod',edgecolor='black')
plt.title('Product and Profit Data')
plt.xlabel('Product')
plt.ylabel('Profit')
plt.show()
```

```python
d=df.groupby('Region')['Profit'].mean()
```

```python
x=d.index
y=d.values
plt.bar(x,y,color='Purple',edgecolor='black')
plt.title('Region and profit Data')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```

```python
d=df.groupby('Region')['Sales'].mean()
```

```python
x=d.index
y=d.values
plt.bar(x,y,color='coral',edgecolor='black')
plt.title('Region and Sales Data')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
d=df.groupby('Category')['Profit'].mean()
d
```

```python
y=d.values
x=d.index
plt.bar(x,y,color='skyblue',edgecolor='black')
plt.title('Category and Profit Data')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

#### histogram

```python
data=df['Profit']
plt.hist(data,color='grey',edgecolor='black')
plt.title('Profit of products')
plt.xlabel('Profit')
plt.show()
```

#### piechart

```python
d=df.groupby('Product')['Sales'].sum().reset_index()
d
```

```python
x=d['Sales']
mylabels=d['Product']
mycolors=["blue","skyblue","powderblue","cyan"]
plt.pie(x,labels=mylabels,autopct='%1.1f%%',colors=mycolors,shadow=True)
plt.legend(title='Product and Sales Data',loc='upper left',bbox_to_anchor=(1.0,1))
plt.xlabel('Sales For each product')
plt.show()
```

```python
d=df.groupby('Category')['Profit'].sum().reset_index()
d
```

```python
x=d['Profit']
mylabels=d['Category']
mycolors=["goldenrod","gold","yellow"]
plt.pie(x,labels=mylabels,autopct='%1.1f%%',colors=mycolors,shadow=True)
plt.legend(title='category and Profit Data',loc='upper left',bbox_to_anchor=(1.0,1))
plt.xlabel('profit for each category')
plt.show()
```

```python
d=df.groupby('Region')['Profit'].sum().reset_index()
d
```

```python
x=d['Profit']
mylabels=d['Region']
mycolors=["Green","lightgreen","lightseagreen","mediumaquamarine"]
plt.pie(x,labels=mylabels,autopct='%1.1f%%',colors=mycolors,shadow=True)
plt.legend(title='category and Profit Data',loc='upper left',bbox_to_anchor=(1.0,1))
plt.xlabel('profit for each region')
plt.show()
```
